```python
import unicodedata

def capitalize_and_remove_accents(word):
  """
  Capitalize the first letter and remove accents from a word.

  Args:
    word (str): The word to be processed.

  Returns:
    str: The processed word with the first letter capitalized and accents removed.
  """
  # Remove accents
  word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8')
  # Capitalize the first letter
  word = word.upper()
  return word
```

```python
name_cities_gdf_cap = [capitalize_and_remove_accents(w) for w in name_cities_gdf]
name_cities_gdf_cap
```
